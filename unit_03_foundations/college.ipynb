```python
## Import pandas and matplotlib (remember "% matplotlib inline")
```

```python
%matplotlib inline
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

#import os
#print (os.getcwd())
#print (os.listdir())
```

```python
## Read in the data and explore the head and tail
```

```python
data = pd.read_csv('recent-grads.csv')
print (data.describe())
data.head()
data.tail()
```

```python
## Drop any rows with missing values
```

```python
data_discard = data.loc[(np.isnan(data['Unemployment_rate'])),data.columns]
data_discard.describe()

data = data[(np.isnan(data['Unemployment_rate'])==False)]
data.describe()
# simplier way:
# data = data.dropna()
```

```python
## Make a scatter plot of sample size vs. median
```

```python
# native pandas
#data.plot.scatter(x=['Sample_size'], y=['Median'])
# Median.min = 22000
# Median.max = 110000
# Sample.min = 2
# Sample.max = 4212
plt.scatter(x=data['Sample_size'], y=data['Median'])
plt.ylim(ymin=20000)
plt.xlim(xmin=0)
plt.show()
```

```python
## Make a scatter plot of sample size vs. unemployment rate
```

```python
# Unemployment.min = 0.0
# Unemployment.max = 0.177226
plt.scatter(x=data['Sample_size'], y=data['Unemployment_rate'])
plt.ylim(ymin=-0.01, ymax=0.2)
plt.xlim(xmin=0)
plt.show()
```

```python
## Make a scatter plot of full time vs. median
```

```python
# Fulltime.min = 111
# fulltime.max = 251540
plt.scatter(x=data['Full_time'], y=data['Median'])
plt.xlim(xmin=0)
plt.ylim(ymin=20000)
plt.show()
```

```python
## Make a scatter plot of men vs. median
```

```python
# Men.min = 119
# Men.max = 173809
plt.scatter(x=data['Men'], y=data['Median'])
plt.xlim(xmin=0)
plt.ylim(ymin=20000)
plt.show()
```

```python
## Make a scatter plot of women vs. median
```

```python
# Women.min = 0
# Women.max = 307087
plt.scatter(x=data['Women'], y=data['Median'])
plt.xlim(xmin=0)
plt.ylim(ymin=20000)
plt.show()
```

```python
## Create a for loop that plots the histogram of sample size, median, employed, full time, unemployment rate, 
## men and women (remember to use the exact column titles)
```

```python
cols = ['Sample_size','Median','Employed','Full_time','Unemployment_rate','Men','Women']
#cols = ['Sample_size']
for col in cols:
    plt.hist(data[col])
    plt.title(col)
    plt.ylabel('Frequency')
    plt.xlabel(col + ' value')
    plt.show()
```
