## Exercise 5 - Find the global maximum and minimum for the following equation.

    f(x) = x^5 - 30(x)^3 + 50(x), where -5 <= x <= 5

```python
expr = x**5 - 30*(x**3) + 50*x
expr_der = sy.diff(expr, x)

print ('In order to find absolute min/max valies of a continuous function on a closed interval [a, b]:')
print (' - find the value of f at the critical numbers of f in (a, b) (i.e. points where derivative=0)')
print (' - find the value of f at endpoints')
print ('The largest is the absolute max, and the smallest is the absolute min.\n')
extremes = {}
extremes[5] = expr.evalf(subs={x: 5})
extremes[-5] = expr.evalf(subs={x: -5})
for i in sy.solveset(expr_der, x):
    if i.evalf() >-5 and i.evalf()<5:
        extremes[expr.evalf(subs={x: i})] = i

print ('Global max is (x = %f, y = %f)' % (extremes[max(extremes.keys())].evalf(), max(extremes.keys())))
print ('Global min is (x = %f, y = %f)' % (extremes[min(extremes.keys())].evalf(), min(extremes.keys())))

from sympy.plotting import plot
p1 = plot(expr, show=False, xlim=(-5.5,5.5), ylim=(-730,730), line_color='b')
p2 = plot(expr_der, show=False, xlim=(-5.5,5.5), ylim=(-730,730), line_color='grey')
p1.extend(p2)
p1.show()
```
